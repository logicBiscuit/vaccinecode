```python
class Employee:
    
    raise_amt = 1.12
    def __init__(self,first,last,pay):
        self.first = first
        self.last = last 
        self.pay = pay
        self.email = first + '.' +last +'@company.com'
        
        
    def full_name(self):
        return('{} {}'.format(self.first,self.last))
    
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amt)
    
    @classmethod
        
        
emp_1 = Employee('sonu','veli',3500)
emp_2 = Employee('sunny','woof',1200)

emp_1.apply_raise()
print(emp_2.raise_amt)
print(emp_1.pay)



# print(emp_1.email)

# print(emp_2.full_name()) 
# print(emp_1.)
```

```python
class food:
    def __init__(self, name,proteins,fat,carbohydrates):
        self.name = name
        self.proteins = proteins
        self.fat = fat
        self.carbohydrates = carbohydrates
        self.energy = proteins + fat + carbohydrates 
    
class nutrition:

    def __init__(self,s,lowl,upl ):
        self.s = s
        self.lowl = lowl
        self.upl =upl

class execute:
    def __init__(self,elist,llist,reqene):
        self.elist = elist
        self.llist = llist
    
           
    def setEnergy(self):
        print("Energy of Food:")
        for i in self.elist:
            for j in self.llist:
                if (i.energy<j.upl and i.energy>j.lowl):
                    print(i.name,"-",i.energy,"-",j.s)
    
    def matchcriteria(self):
        f1 = {}
        for i in self.elist:
            if(i.energy<=reqene):
                f1[i.name] = i.energy
        if(f1):
            print("Food within given criteria:")
            for k in f1:
                print(k,":",f1[k])
        else :
            print("No Food Found")
        
                
        
            
            
        
            
            
if __name__=="__main__":
    no = int(input())
    val = []
    for i in range(no):
        name = input()
        pro = float(input())
        fat = float(input())
        carb = float(input())
        val.append(food(name,pro,fat,carb))
    enegval  = int(input())
    enelis = []
    for i in range(enegval):
        enetype = input()
        lowl = int(input())
        upl = int(input())
        enelis.append(nutrition(enetype,lowl,upl))
    
    reqene = int(input())
    obj = execute(val , enelis, reqene)
    obj.setEnergy()
    obj.matchcriteria()
```
